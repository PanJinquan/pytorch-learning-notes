```python
%matplotlib notebook
import tensorwatch as tw
# 连接执行tensorwatch的终端w.observe(step=iteration, loss=loss, face_loss=face_loss,body_loss=head_loss)
client = tw.WatcherClient()
# 迭代次数和训练的正确率之间的关系
# 使用lambda获取终端中的参数
loss_stream = client.create_stream(expr='lambda d:(d.step, d.loss)')
# 可视化设置
loss_plot = tw.Visualizer(loss_stream, vis_type='line', xtitle='step', ytitle='Train Loss')
loss_plot.show()
# 迭代次数和训练的正确率之间的关系
face_loss = client.create_stream(expr='lambda d:(d.step, d.face_loss)')
face_plot = tw.Visualizer(face_loss, vis_type='line', xtitle='step', ytitle='face_loss')
face_plot.show()

body_loss = client.create_stream(expr='lambda d:(d.step, d.body_loss)')
body_plot = tw.Visualizer(body_loss, vis_type='line', host=face_plot, xtitle='step', ytitle='body_loss', yrange=(0,))
body_plot.show()
```
