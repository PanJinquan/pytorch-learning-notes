```python
%matplotlib notebook
import tensorwatch as tw
# 连接执行tensorwatch的终端
client = tw.WatcherClient()
# 迭代次数和训练的正确率之间的关系
# 使用lambda获取终端中的参数
loss_stream = client.create_stream(expr='lambda d:(d.step, d.loss)')
# 可视化设置
loss_plot = tw.Visualizer(loss_stream, vis_type='line', xtitle='Epoch', ytitle='Train Loss')
loss_plot.show()
# 迭代次数和训练的正确率之间的关系
acc_stream = client.create_stream(expr='lambda d:(d.step, d.train_accuracy)')
acc_plot = tw.Visualizer(acc_stream, vis_type='line', host=loss_plot, xtitle='Epoch', ytitle='Train Accuracy', yrange=(0,))
acc_plot.show()
```
